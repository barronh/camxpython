
# Get Examples Folder

* Navigate to the tutorial http://github.com/barronh/camxpython
* Go to the documentation.
* Find "Download all examples in Python source code: auto_examples_python.zip" and download it.
* Make a `~/examples folder` and copy the `auto_examples_python.zip` file there.
* Unzip the `auto_examples_python.zip` (`unzip auto_examples_python.zip`).

When you're done, run `tree -L 2 ~/examples`. You should get something like
